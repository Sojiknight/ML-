```python
adm = pd.read_csv('Admission_Predict_Ver1.1.csv')
```

```python
adm['Chance of Admit ']
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
        
    else:
        return Age
```

```python
def adm_switch(x):
       
    if x in adm['Chance of Admit '] <= 0.49:
        return 0
    else:
        return 1
```

```python
adm['adm_switch'] = adm['Chance of Admit '].apply(adm_switch)
```
